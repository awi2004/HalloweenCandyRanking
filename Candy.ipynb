# Data Visualisation & Descriptive analytics

```python
def visualisation(titleViolin,X_violin,Y_violin,X_labelVio,titlePie,pieValue,titleBoxPlot,y_box,x_box,
                  xLable_box,label):   
    '''
     function returns the plot of all the content of Candy with winpercent, pricepercent, its overall contribution in making candy.
    '''
    plt.figure(figsize=(30,10))
    

    plt.subplot(1,3,1)
    plt.title(titleViolin,fontsize=16)
    sns.violinplot(data=data,x=X_violin,y=Y_violin,color="brown",palette="husl",hue=X_violin)
    plt.xlabel(X_labelVio)

    plt.subplot(1,3,3)
    plt.title(titlePie,fontsize=16)
    plt.pie(pieValue, autopct='%1.1f%%',colors=["#4ECDC4","#FF6B6B"],labels=["Not have "+label,"Have "+ label]
            ,shadow=True, startangle=90)

    plt.subplot(1,3,2)
    plt.title(titleBoxPlot,fontsize=16)
    sns.boxplot(y=y_box, x=x_box, data=data, palette="gist_stern_r",hue=x_box)
    plt.xlabel(xLable_box)
    plt.savefig(label)
    plt.show()
```

# Data preparation and modelling

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, Normalizer
from sklearn.metrics import r2_score
X=data[data.columns[1:-1]]
sc=StandardScaler()
X=sc.fit_transform(X)
y=data[data.columns[-1]] #winpercent as label(regressor problem)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import SGDRegressor
from xgboost import XGBRegressor
import math
#create a list of classifier to train the data on them
def model(xtrain, ytrain):
    '''
    #Function model the data on DecisionTreeRegressor, RF Regressor, SGDRegressor and XGB Regressor.
    #param: features and labels.
    '''
    rmse_err = []
    classifiers = [
        DecisionTreeRegressor(max_depth=6),
        RandomForestRegressor(n_estimators=200),
        SGDRegressor(max_iter=1000),
        XGBRegressor(max_depth=2,n_estimators=100,min_child_weight=8)
        
        ]

    for classifier in classifiers:#range(1,6)
        #rmse_err = []
        pl_classifier = Pipeline([
                 #('best_feat', SelectKBest(k=ki)),
                 ('clf', classifier)
                ])
       
        #scores = cross_val_score(pl_classifier, xtrain, ytrain, cv=2)
       
        classifier.fit(xtrain,ytrain)   
        y_pred=classifier.predict(X_test)
        scores=r2_score(y_test,y_pred)
        rmse_err.append(math.sqrt(mean_squared_error(y_test,y_pred)))
        #rmse_err.append(scores)
        print("Ensemble | k %s | r2_score: %0.6f  Rmse_err score: %0.6f" % (classifier, scores, rmse_err[-1]))
```

```python
model(X_train,y_train)
```

```python
from itertools import cycle, islice
models = ['Decision Tree','RandomForest','SGD','XGBRegressor']
my_colors = list(islice(cycle(['b', 'r', 'g', 'y', 'k']), None, 4))
pd.DataFrame.from_dict({'Name':models,'RMSE':rmse_err}).sort_values(by='RMSE',ascending=False).plot(x='Name',y='RMSE',kind='barh',color=my_colors,sort_columns=True,figsize = (8,5),legend=False,title='Performance of various Regression based algos')
```

Inference from modelling:
1. XGB Regressor works best for modelling this data set.
2. On training the data on 75% of the record and tested on 25% of records, best r2Sqaure Score i.e (1 - residual sum of square / total sum of squares) is 0.36 and lowest Rmse_error is 11.16 from XGB Regressor.
3. Above plot shows perfomance of Regrssors.
